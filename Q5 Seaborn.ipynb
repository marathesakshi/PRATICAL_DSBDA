```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Example data
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [2, 3, 5, 7, 11]
}
```

```python
# Convert data to DataFrame
df = pd.DataFrame(data)
```

```python
# a. Line Plot
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='X', y='Y')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# b. Bar Plot
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='X', y='Y')
plt.title('Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(axis='y')
plt.show()
```

```python
# c. Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='X', y='Y')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
