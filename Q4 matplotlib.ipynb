```python
import matplotlib.pyplot as plt
```

```python
# Example data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
```

```python
# a. Line Plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', color='b', linestyle='-')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# b. Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(x, y, color='g')
plt.title('Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(axis='y')
plt.show()
```

```python
# c. Pie Plot
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [20, 30, 25, 15, 10]
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice
```

```python
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Plot')
plt.show()
```
